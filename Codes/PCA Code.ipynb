```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

eigenvec_file = 'C:/Users/mohak/Desktop/rice_pca.eigenvec'

try:
    df_pca = pd.read_csv(eigenvec_file, sep='\s+', header=None)
    df_pca.rename(columns={0: 'FID', 1: 'IID'}, inplace=True)
    pc_cols = [f'PC{i}' for i in range(1, df_pca.shape[1] - 1)]
    df_pca.columns = ['FID', 'IID'] + pc_cols

    plt.figure(figsize=(10, 8))
    sns.scatterplot(x='PC1', y='PC2', data=df_pca, alpha=0.7)
    plt.title('PCA Plot: PC1 vs PC2')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.savefig('C:/Users/mohak/Desktop/pca_plot_pc1_vs_pc2.png')
    plt.close()
    print("PCA plot 'pca_plot_pc1_vs_pc2.png' generated.")

except FileNotFoundError:
    print(f"Error: The file '{eigenvec_file}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
